# Bibliotecas e func "newCol"


```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# FUNCAO NOVA COLUNA COM SWC_PARAMS
def newCol(df):
    df['nome'] = 'NaN'
    for i, linha in df.iterrows():
        if linha['tipo'] == 0:
            df.loc[i, 'nome'] = 'undefined'
        elif linha['tipo'] == 1:
            df.loc[i, 'nome'] = 'soma'
        elif linha['tipo'] == 2:
            df.loc[i, 'nome'] = 'axon'
        elif linha['tipo'] == 3:
            df.loc[i, 'nome'] = 'basal dendrite'
        elif linha['tipo'] == 4:
            df.loc[i, 'nome'] = 'apical dendrite'
        elif linha['tipo'] == 5:
            df.loc[i, 'nome'] = 'custom'
        elif linha['tipo'] == 6:
            df.loc[i, 'nome'] = 'unspecified neurites'
        elif linha['tipo'] == 7:
            df.loc[i, 'nome'] = 'glia processes'
```

```python
tab = pd.read_excel('Neuronio_1.xlsx')
newCol(tab)
tab.head()
```

```python
#CRIACAO MATRIZ DE ADJACENCIA
index = np.arange(0, 1725)
tab_adj = tab[['knot', 'pai']]
tab_adj['knot'] = index 
tab_adj['pai'] = tab_adj['pai'] - 1
tab_adj
```

```python
#TABELA SOMENTE COM OS ATRIBUTOS DE CADA PARTE DO NEURONIO
tab_atributos = tab[['tipo','x', 'y', 'z','raio', 'nome']]

atributos = tab_atributos.to_dict(orient='index')
tab_atributos
```

```python
#CRIACAO DA REDE NEURAL
G = nx.from_pandas_edgelist(tab_adj, source='knot', target='pai')
```

```python
#TABELA DE POSICOES NECESSARIA PARA PLOTAR A REDE COM O FORMATO DO NEURONIO
new_index = np.arange(1,1726)
tab_pos = tab[['x', 'y']]
tab_pos.index = new_index
pos = tab_pos.to_numpy()
```

```python
#PLOT DA REDE NEURAL 
options = {
    'edge_color': 'black',
    'node_size': 2,
    'width': 1,
}

plt.figure(figsize=(10, 12))
nx.draw_networkx(G, pos,  node_size=1, width=1, edge_color='black', with_labels=False)
plt.show()
```

```python
#SET DOS ATRIBUTOS DE CADA NOH
nx.set_node_attributes(G, atributos)

#ARMAZENANDO O ATRIBUTO DE CADA NOH EM UM DICT
pos_x = nx.get_node_attributes(G, 'x')
pos_y = nx.get_node_attributes(G, 'y')
pos_z = nx.get_node_attributes(G, 'z')
tipos = nx.get_node_attributes(G, 'tipo')
nomes = nx.get_node_attributes(G, 'nome')
```

```python
def make_img(df, res=1):
    
    df = tab[['tipo', 'x', 'y']]
    df = df.astype({"tipo": np.uint8, "x": int, "y": int})
    
    #definicao da resolucao (res=1 se nao especificado na chamada da funcao)
    df['x'] = df['x']/res
    df['y'] = df['y']/res
    df = df.astype({"x": int, "y": int})
    
    #shift da imagem (menor valor ser o '0')
    shift_x = min(df['x'])
    shift_y = min(df['y'])
    df['x'] = df['x'] - shift_x
    df['y'] = df['y'] - shift_y
    
    #Rotacao de 90 graus
    df['y'] = max(df['y']) - df['y']
    
    #Definicao dos tons de cinza de cada pixel de acordo com seu tipo (soma, axon, dendrite)
    df.loc[df.tipo == 1,'tipo']= 255
    df.loc[df.tipo == 2,'tipo']= 128
    df.loc[df.tipo == 3,'tipo']= 255
    df.loc[df.tipo == 4,'tipo']= 255
    
    #CRIANDO A MATRIZ 'IMAGEM'
    
    #criacao da matriz de zeros
    img = np.zeros([max(df['y']) + 1,max(df['x']) + 1], dtype=np.uint8)
    
    #indexacao dos pixels
    for index, row in df.iterrows():
        img[row['y'], row['x']] = row['tipo']
        
    #inversao da imagem
    img = 255 - img
    
    return img
```

```python
%matplotlib notebook

img = make_img(tab)

plt.figure(figsize=(12, 6))
plt.imshow(img, cmap='gray')

plt.show()
```
