```python
import pandas as pd
import numpy as np
import re
```

```python
def read_swc(arquivo):
    
    df = pd.read_csv(arquivo, header=None)
    display(df)
    colunas = ['knot','tipo','x', 'y','z', 'raio','pai']
    remove = []
    for i, row in df.iterrows():
        linha = row[0].strip()
        if linha[0] == '#':
            remove.append(i)
        if linha[0] == '1':
            break
    
    df = pd.read_csv(arquivo, skiprows=remove, header=None, sep=' ')
    df.drop([0],axis=1, inplace=True) #Tira a primera coluna quando tem um espaco
    df.columns = colunas
    df.loc[0, 'pai'] = 1
    
    return df

df = read_swc(arquivo='neuron-1.swc')
display(df)
```

```python
df_mouse = pd.read_csv('conjunto-1-mouse-DSA.csv')
display(df_mouse.loc[8470])
df_mouse = df_mouse[['N_stems','N_bifs', 'N_branch', 'Length']]
df_mouse = df_mouse.astype({"N_stems": float, "N_bifs": float, "N_branch": float})
means = df_mouse.mean()
print(f'Media das medidas: \n{means}\n')

# Pega a menor diferenca entre respectivo atributo e a media geral do atributo

distancias = []
for index, linha in df_mouse.iterrows():
    
    dist = np.sqrt((means['N_stems'] - linha['N_stems'])**2 + (means['N_branch'] - linha['N_branch'])**2 + 
                   (means['N_bifs'] - linha['N_bifs'])**2 + (means['Length'] - linha['Length'])**2)
    
    distancias.append(dist)

    
neuron = distancias.index(min(distancias))
indices = np.argsort(distancias)
df_mouse.iloc[indices[1]]
```
